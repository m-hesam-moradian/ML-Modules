```python
import pandas as pd

xls = pd.ExcelFile("data/85_Soccer_ETR_LGBR_COA_BWO.xlsx")

df = pd.read_excel("data/85_Soccer_ETR_LGBR_COA_BWO.xlsx", sheet_name="DATA")

y = df["markat value"]

X = df.drop(columns=["markat value"])

import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(X):
    vif_data = pd.DataFrame()
    vif_data["feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data

def iterative_vif(X, threshold=1.0):
    X_current = X.copy()
    tables = [] 

    while True:
        vif = calculate_vif(X_current)
        tables.append(vif)  

        max_vif = vif["VIF"].max()
        if max_vif <= threshold:
            break

       
        max_vif_feature = vif.loc[vif["VIF"].idxmax(), "feature"]
        print(f"Removing '{max_vif_feature}' with VIF={max_vif:.2f}")
        X_current = X_current.drop(columns=[max_vif_feature])

      
        if X_current.shape[1] == 0:
            print("All features removed.")
            break

    return tables, X_current



vif_tables, X_filtered = iterative_vif(X, threshold=1.0)

for i, table in enumerate(vif_tables):
    print(f"\nIteration {i+1} VIF values:")
    print(table)

```
